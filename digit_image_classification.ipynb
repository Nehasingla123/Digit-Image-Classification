```python
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
```

```python
digits=load_digits()
```

```python
print(dir(digits))
```

```python
digits.data[0]
```

```python
plt.gray()
plt.matshow(digits.images[0])
```

```python
plt.gray()
for i in range(5):
    plt.matshow(digits.images[i])
```

```python
digits.target[0:5]
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(digits.data,digits.target,test_size=0.2)
Y_test
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
model.score(X_test,Y_test)
```

```python
plt.matshow(digits.images[1000])
```

```python
digits.target[1000]
```

```python
model.predict([digits.data[1000]])
```

```python
y_predicted=model.predict(X_test)
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test,y_predicted)
print(cm)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 7))
plt.scatter(y_predicted, Y_test)
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

# Assuming cm is a confusion matrix (generated using sklearn's confusion_matrix function)
# And Y_test contains the actual values, and y_pred contains predicted values

plt.figure(figsize=(10, 7))

# Plot the confusion matrix using a heatmap
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')

# Label the axes
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')

# Show the plot
plt.show()
```
